```python
# 5: using RESNET Model


# ==========================================================
#  RESNET18 + CLINICAL  •  OLIVES DATASET  (train / val / test)
# ==========================================================


# -------- 2.  LOAD OLIVES DATA --------------------------------------
ds = load_dataset("gOLIVES/OLIVES_Dataset", "biomarker_detection")
train_split = ds["train"]              # 3 556 samples
test_split  = ds["test"]               # 1 186 samples

# reserve part of train_split for validation
VAL_SIZE = 500                         # ← change if you like
TRAIN_SIZE = len(train_split) - VAL_SIZE

train_hf = train_split.select(range(TRAIN_SIZE))
val_hf   = train_split.select(range(TRAIN_SIZE, len(train_split)))

# -------- 3.  CUSTOM TORCH DATASET ----------------------------------
class OLIVESDataset(Dataset):
    def __init__(self, hf_split):
        self.data = hf_split
        self.tfm  = transforms.Compose([
            transforms.Resize((32, 32)),   # 64 64
            transforms.Grayscale(num_output_channels=3),
            transforms.ToTensor(),
        ])

    def __len__(self): return len(self.data)

    def __getitem__(self, idx):
        ex = self.data[idx]

        labels = torch.tensor(
            [float(ex[f"B{i}"] or 0.0) for i in range(1, 7)], dtype=torch.float32
        )
        clinical = torch.tensor(
            [float(ex["BCVA"] or 0.0), float(ex["CST"] or 0.0)], dtype=torch.float32
        )

        img = ex["Image"] if isinstance(ex["Image"], Image.Image) \
             else Image.open(ex["Image"]).convert("L")
        oct_tensor = self.tfm(img)

        return {"oct": oct_tensor, "clinical": clinical, "label": labels}

# build PyTorch datasets / loaders
train_ds = OLIVESDataset(train_hf)
val_ds   = OLIVESDataset(val_hf)
test_ds  = OLIVESDataset(test_split)

train_loader = DataLoader(train_ds, batch_size=256, shuffle=True)
val_loader   = DataLoader(val_ds,   batch_size=32, shuffle=False)
test_loader  = DataLoader(test_ds,  batch_size=32, shuffle=False)

# -------- 4.  MODEL --------------------------------------------------
class MultiModalNet(nn.Module):
    def __init__(self):
        super().__init__()
        backbone = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
        backbone.fc = nn.Identity()          # 512‑dim
        self.oct_model = backbone
        self.head = nn.Sequential(
            nn.Linear(512 + 2, 128),
            nn.ReLU(),
            nn.Linear(128, 6),
            nn.Sigmoid()
        )

    def forward(self, img, clin):
        feat = self.oct_model(img)
        x = torch.cat([feat, clin], dim=1)
        return self.head(x)

model = MultiModalNet().to(device)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# -------- 5.  TRAIN / VALIDATE --------------------------------------
EPOCHS = 5
train_loss = []   # holds average train loss per epoch
val_loss   = []   # holds average val   loss per epoch
for epoch in range(1, EPOCHS + 1):
    # --- train ---
    model.train()
    epoch_train_loss = 0.0 # Initialize an accumulator for the epoch's loss
    for batch in tqdm(train_loader, desc=f"Train {epoch}/{EPOCHS}"):
        img = batch["oct"].to(device)
        cli = batch["clinical"].to(device)
        lab = batch["label"].to(device)

        optimizer.zero_grad()
        loss = criterion(model(img, cli), lab)
        loss.backward()
        optimizer.step()
        epoch_train_loss += loss.item() # Accumulate batch loss

    epoch_train_loss /= len(train_loader) # Calculate average epoch loss

    # --- validate ---
    model.eval()
    epoch_val_loss = 0.0 # Similar accumulator for validation loss
    with torch.no_grad():
        for batch in val_loader:
            img = batch["oct"].to(device)
            cli = batch["clinical"].to(device)
            lab = batch["label"].to(device)
            epoch_val_loss += criterion(model(img, cli), lab).item()
    epoch_val_loss /= len(val_loader)


    train_loss.append(epoch_train_loss) # Append average epoch losses to the list
    val_loss.append(epoch_val_loss)

    print(f"Epoch {epoch:02d} | Train loss {epoch_train_loss:.4f} | Val loss {epoch_val_loss:.4f}")

# -------- 6.  TEST ---------------------------------------------------
model.eval()
all_preds, all_labels = [], []
with torch.no_grad():
    for batch in tqdm(test_loader, desc="Test"):
        img = batch["oct"].to(device)
        cli = batch["clinical"].to(device)
        preds = model(img, cli).cpu()
        all_preds.append(preds)
        all_labels.append(batch["label"])

all_preds  = torch.cat(all_preds).numpy()
all_labels = torch.cat(all_labels).numpy()
print("✅ Evaluation on full test set complete.")

```

```python
import matplotlib.pyplot as plt

epochs = range(1, len(train_loss) + 1)

plt.figure(figsize=(8, 5))
plt.plot(epochs, train_loss, marker='o', label='Train loss')
plt.plot(epochs, val_loss,   marker='s', label='Val loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss per Epoch')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
